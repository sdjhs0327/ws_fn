```python
from sympy import Symbol, solveset, Interval
import numpy as np
```

```python
def cal_ret(k, s, c, t):
    '''
    * k: 목표자산, s: 시드머니, c: 월투자금, t: 투자기간(연)
    * r: 목표수익률
    '''
    k = k*1e4
    s = s*1e4
    c = c*1e4*12
    if k <= s+c*t:
        return 0
    else:        
        r=Symbol('r')
        equation = s*((1+r)**(t)) + c*((1-((1+r)**(t)))/(1-(1+r))) - k
        val = solveset(equation, r, Interval(0, 1))
        solve_r = float(list(val)[0])

        return (np.round(solve_r, 6))*100
```

```python
k, s, c, t = 50000, 1000, 200, 10
```

```python
demanded_return = cal_ret(k, s, c, t)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 예제
x = np.linspace(0, 10, 100)
y1 = np.sin(x)  # 메인 데이터 (오렌지)
y2 = np.cos(x)  # 보조 데이터 1 (블루)
y3 = np.tan(x) / 10  # 보조 데이터 2 (그린)

# 색상 설정
main_color = "#E67E22"  # 부드러운 오렌지
red_color = "#C0392B"   # 세련된 레드
blue_color = "#2980B9"  # 차분한 블루
green_color = "#27AE60" # 긍정적 데이터

# 그래프 그리기
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y1, color=main_color, linewidth=2, label="메인 데이터 (오렌지)")
ax.plot(x, y2, color=red_color, linestyle="--", linewidth=1.5, label="보조 데이터 1 (블루)")
ax.plot(x, y3, color=blue_color, linestyle=":", linewidth=1.5, label="보조 데이터 2 (그린)")

ax.set_title("📊 개선된 그래프 색 조합", fontsize=16, fontweight="bold")
ax.set_xlabel("X 축", fontsize=12)
ax.set_ylabel("Y 축", fontsize=12)
ax.legend()
ax.grid(True, linestyle="--", alpha=0.7)

plt.show()
```
