```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import datetime as dt
from dateutil.relativedelta import relativedelta
import math

##절대경로 참조
import sys, os
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname('__file__'))))
from library.quant_functions import anal_funcs, data_funcs

import yfinance as yf
```

```python
path = 'D:/data/stock/'
```

```python
mylist = pd.read_csv(f"{path}list.csv")
```

```python
save_date = dt.datetime.now().strftime(format='%Y%m%d')
save_weekday = dt.datetime.now().weekday()
try:
    ## 저장요일
    os.mkdir(f"{path}data/{save_weekday}")
except:
    pass

try:
    ## 저장요일
    os.mkdir(f"{path}hdata/{save_weekday}")
except:
    pass
```

```python
for ticker in mylist['Ticker']:
    try:
        data = yf.download(ticker, interval='1m')
        if len(data) > 1:
            try:
                ## Ticker
                os.mkdir(f"{path}data/{save_weekday}/{ticker}")
            except:
                print('이미 경로가 존재합니다.')
                pass
            data.to_csv(f"{path}data/{save_weekday}/{ticker}/{save_date}.csv")
    except:
        print('저장할 수 없습니다.')
        pass
```

```python
for ticker in mylist['Ticker']:
    try:
        data = yf.download(ticker, interval='1h', start=dt.datetime.now() - relativedelta(years=2))
        if len(data) > 1:
            try:
                ## Ticker
                os.mkdir(f"{path}hdata/{save_weekday}/{ticker}")
            except:
                print('이미 경로가 존재합니다.')
                pass
            data.to_csv(f"{path}hdata/{save_weekday}/{ticker}/{save_date}.csv")
    except:
        print('저장할 수 없습니다.')
        pass
```
